## Gráficos semilogarítmicos con semilogx() y semilogy(): 

```python
import matplotlib.pyplot as plt
plt.semilogy?
```

### Gráfico semilogarítmico en el eje y:

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(1,100,10)
plt.semilogy(x,np.exp(x),'r--',base=5); #base: base del logarítmico
plt.grid();
```

### Gráfico semilogarítmico en el eje x:

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(1,1000,100)
plt.semilogx(x,np.log(x),'y.');
plt.grid();
```

Se pueden combinar ambas funciones:

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(1,100,10)
y=x**2
plt.semilogy(x,np.exp(x),'r--',base=5)
plt.semilogx(base=2);
```

## Gráficos logarítmicos con loglog():

```python
import matplotlib.pyplot as plt
plt.loglog?
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0.1,16.5*np.pi,200)
y=np.sin(x)**2
fig=plt.figure(figsize=(8,6))
plt.loglog(x,y,'r.')
plt.loglog(x,y,'g',base=3);
```
