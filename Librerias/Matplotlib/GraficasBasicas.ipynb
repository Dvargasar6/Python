## Gr치ficas r치pidas sin usar fig, ax:

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
plt.plot(x,y,'m');
plt.plot(x,z,'g*');
plt.grid()
plt.ylabel('Eje y');
plt.xlabel('Eje x')
plt.title('Gr치fico');
```

## Gr치ficas usando fig,ax:

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,2*np.pi,100)
y1=np.sin(x)
y2=np.cos(x)
fig, ax = plt.subplots(figsize=(10,5)) #tama침o de la figura
ax.plot(x,y1,'--r',label='sin(x)') #eje x, eje y, tipo de l칤nea y leyenda 
ax.plot(x,y2,'.b',label='cos(x)')
ax.legend()  #imprimir leyenda en el gr치fico
ax.grid()
ax.set_title("Funciones") #T칤tulo
ax.set_xlabel("x")  #T칤tulos ejes
ax.set_ylabel("y");
#fig.savefig("migr치fico.png")   #Guardar la imagen en png
```
