## Gráficas rápidas sin usar fig, ax:

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
plt.plot(x,y,'m');
plt.plot(x,z,'g*');
plt.grid()
plt.ylabel('Eje y');
plt.xlabel('Eje x')
plt.title('Gráfico');
```

## Gráficas usando fig,ax:

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,2*np.pi,100)
y1=np.sin(x)
y2=np.cos(x)
fig, ax = plt.subplots(figsize=(10,5)) #tamaño de la figura
ax.plot(x,y1,'--r',label='sin(x)') #eje x, eje y, tipo de línea y leyenda 
ax.plot(x,y2,'.b',label='cos(x)')
ax.legend()  #imprimir leyenda en el gráfico
ax.grid()
ax.set_title("Funciones") #Título
ax.set_xlabel("x")  #Títulos ejes
ax.set_ylabel("y");
#fig.savefig("migráfico.png")   #Guardar la imagen en png
```
