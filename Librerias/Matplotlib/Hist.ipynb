Esta función permite realizar histogramas.

```python
import matplotlib.pyplot as plt
import numpy as np
X=np.random.random(500)
plt.hist(X);
```

```python
X=np.random.normal(165,2000,100000)
plt.hist(X,bins=100,alpha=0.6,color='y');
plt.title('Distribución normal',fontsize=20,color='r');
plt.grid()
```
