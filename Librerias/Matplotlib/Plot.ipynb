## FMT:

Los parámetros fmt sirven para definir color, estilo de línea o marcador de una forma simple.

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
plt.plot(x,y,'m--');
plt.plot(x,z,'b4');   #Es posible gráficar múltiples relaciones.
#plt.plot(x,y,'m--',x,z,'g*'); Alternativa
plt.grid()
plt.ylabel('Eje y');
plt.xlabel('Eje x')
plt.title('Gráfico');
```

## Line2D:

Es una forma de modificar la apariencia de las gráficas por medio de palabras clave: \
*color*: Color\
*marker*: Marcador\
*linestyle*: Estilo de línea\
*linewidth*: Ancho de linea\
*markersize*: Tamaño de marcador\
.\
.\
.

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
plt.plot(x,y,color='black',linestyle='dashed',marker='x',markersize=10);
plt.plot(x,z,'g',linewidth=10, linestyle='dashed');
plt.grid()
```
