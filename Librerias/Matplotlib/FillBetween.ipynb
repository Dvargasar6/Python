Esta función permite graficar el área entre dos curvas y=f(x)

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(0,20,0.1)
y1 = np.sin(x)
y2 = x*np.sin(x)
plt.fill_between( x, y1, y2,alpha=0.3, color='m' );
```

# Matplotlib.pyplot.fill_betweenx()

Permite gráficar el área entre dos curvas x=f(y)

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange( 0, 20, 0.1 )
y1 = np.sin(x)
y2 = x
plt.fill_betweenx(x,y1,y2,alpha=0.6,color='g');
```
