# Matplotlib.pyplot.subplots

La función subplots de Matplotlib permite crear matrices de gráficos.\
Subplots se centra en los conceptos de ejes (*axes*) y figura (*figure*).\
Se puede decir que la figura es el lienzo donde se situan los ejes, y los ejes son las coordenadas que usarán las gráficas.

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,2*np.pi,100)
y1=np.sin(x)
y2=np.cos(x)
fig, ax = plt.subplots(2,2) # Una fila, dos columnas
#fig, (ax1,ax2) = plt.subplots(1,2) Alternativa
fig.set_facecolor('y') #Figura color amarillo
ax[0,0].plot(x,y1)
ax[0,1].set_facecolor('b') #Eje 1 color azúl
ax[1,1].set_facecolor('k')
ax[1,1].plot(x,y2,'w--')
ax[1,1].set_xscale('log',base=2) #Cambio de escala
```

### sharex y sharey:

Estos parámetros controlan la compartición de los ejes. Así, sharey = True hace que todos los ejes de la misma fila compartan las mismas ordenadas y sharex = True hace que todos los ejes de la misma columna compartan las mismas abscisas. 

```python
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots(2,2,sharex=True,sharey=True)
```

### Algunos elementos:

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,2*np.pi,100)
y1=np.sin(x)
y2=np.cos(x)
fig, ax = plt.subplots(2,2,figsize=[10,5])
fig.set_edgecolor('y')
ax[0,0].set_title('Gráfica 1') #Título
ax[0,1].grid(); #Cuadrícula
ax[1,0].set_ylabel('y') #Nombres de los ejes
ax[1,0].set_xlabel('x');
ax[1,1].plot(x,y1,label='sin(x)')
ax[1,1].legend(); #Leyenda
```

# Matplotlib.pyplot.subplot

La función subplot de Matplotlib permite agregar ejes a una figura existente. Es una alternativa a subplots a la hora de crear múltiples gráficos.\
También brinda la opción de generar gráficos polares.

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,2*np.pi,100)
y1=np.sin(x)
y2=np.cos(x)
y3=np.exp(x)
plt.figure(figsize=(10,10))
#Primer eje:
plt.subplot(3,2,1,title='Gráfica 1')  # (Filas, Columnas, Posición)
plt.loglog(x,y1);
#Segundo eje:
plt.subplot(322,fc='red') #Alternativa, solo válido para menos de 10 gráficos
plt.plot(x,y2);
plt.ylabel('Y')
#Tercer eje:
plt.subplot(3,2,(3,6),polar=True) #Ocupará desde la posición 3 hasta la 6.
plt.plot(x,y3);
```
