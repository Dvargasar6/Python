La funci칩n subplots de Matplotlib permite crear matrices de gr치ficos.\
Subplots se centra en los conceptos de ejes (*axes*) y figura (*figure*).\
Se puede decir que la figura es el lienzo donde se situan los ejes, y los ejes son las coordenadas que usar치n las gr치ficas.

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,2*np.pi,100)
y1=np.sin(x)
y2=np.cos(x)
fig, ax = plt.subplots(2,2) # Una fila, dos columnas
#fig, (ax1,ax2) = plt.subplots(1,2) Alternativa
fig.set_facecolor('y') #Figura color amarillo
ax[0,0].plot(x,y1)
ax[0,1].set_facecolor('b') #Eje 1 color az칰l
ax[1,1].set_facecolor('k')
ax[1,1].plot(x,y2,'w--')
ax[1,1].set_xscale('log',base=2) #Cambio de escala
```

### sharex y sharey:

Estos par치metros controlan la compartici칩n de los ejes. As칤, sharey = True hace que todos los ejes de la misma fila compartan las mismas ordenadas y sharex = True hace que todos los ejes de la misma columna compartan las mismas abscisas. 

```python
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots(2,2,sharex=True,sharey=True)
```

### Algunos elementos:

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,2*np.pi,100)
y1=np.sin(x)
y2=np.cos(x)
fig, ax = plt.subplots(2,2,figsize=[10,5])
fig.set_edgecolor('y')
ax[0,0].set_title('Gr치fica 1') #T칤tulo
ax[0,1].grid(); #Cuadr칤cula
ax[1,0].set_ylabel('y') #Nombres de los ejes
ax[1,0].set_xlabel('x');
ax[1,1].plot(x,y1,label='sin(x)')
ax[1,1].legend(); #Leyenda
```

La funci칩n subplot de Matplotlib permite agregar ejes a una figura existente. Es una alternativa a subplots a la hora de crear m칰ltiples gr치ficos.\
Tambi칠n brinda la opci칩n de generar gr치ficos polares.

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,2*np.pi,100)
y1=np.sin(x)
y2=np.cos(x)
y3=np.exp(x)
plt.figure(figsize=(10,10))
#Primer eje:
plt.subplot(3,2,1,title='Gr치fica 1')  # (Filas, Columnas, Posici칩n)
plt.loglog(x,y1);
#Segundo eje:
plt.subplot(322,fc='red') #Alternativa, solo v치lido para menos de 10 gr치ficos
plt.plot(x,y2);
plt.ylabel('Y')
#Tercer eje:
plt.subplot(3,2,(3,6),polar=True) #Ocupar치 desde la posici칩n 3 hasta la 6.
plt.plot(x,y3);
```
